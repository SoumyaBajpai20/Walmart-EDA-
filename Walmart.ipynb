```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:\Dataset\Walmart Sales.csv")
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df['amount']=df['Rating']*df['Unit price']
df[['Rating', 'amount']].describe()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.drop(['Invoice ID', 'Date', 'Time'],axis=1,inplace=True)
```

```python
Payment	=df['Payment'].value_counts().reset_index()
Payment	
```

```python
amount=df['amount'].sum()
Quantity=df['Quantity'].sum()

print(f'Total Amount :{round(amount,2)}')
print(f'Total Quantity :{Quantity}')
```

```python
plt.figure(figsize =(4,3))
plt.bar(Payment['Payment'],Payment['count'],color='black',width=0.3)
plt.xlabel('Payment', color='red')
plt.ylabel('Rating',color='red')
plt.grid(color='red')
plt.show()
```

```python
Payment_amount=df.groupby('Payment')[['amount']].sum().reset_index()
Payment_amount
```

```python
plt.figure(figsize =(4,3))
plt.bar(Payment_amount['Payment'],Payment_amount['amount'],color='black',width=0.3)
plt.xlabel('Payment', color='red')
plt.ylabel('amount',color='red')
plt.grid(color='red')
plt.show()
```

```python
amountbyProductline=df.groupby('Product line')[['amount']].sum().reset_index()
amountbyProductline
```

```python
plt.figure(figsize=(5,4))
plt.bar(amountbyProductline['Product line'],amountbyProductline['amount'],color='black')
plt.plot(amountbyProductline['amount'],color='red',marker='.')
plt.xticks(rotation=90)
plt.xlabel('Product', color='red')
plt.ylabel('amount',color='red')
plt.grid()
plt.show()
```

```python
plt.pie(amountbyProductline['amount'])
plt.legend(bbox_to_anchor=(1.05,1.0),labels=amountbyProductline['Product line'])
plt.show()
```

```python
# Payment through Cash
data1=df.where(df['Payment']=='Cash')
data1.dropna(inplace=True)
```

```python
amountByProductline=data1.groupby('Product line')[['Quantity', 'amount']].sum().reset_index()
amountByProductline
```

```python
plt.figure(figsize=(5,4))
plt.bar(amountbyProductline['Product line'],amountbyProductline['amount'],color='black')
plt.plot(amountbyProductline['amount'],color='red',marker='.')
plt.xticks(rotation=90)
plt.xlabel('Product', color='red')
plt.ylabel('amount',color='red')
plt.grid()
plt.show()
```

```python
# Payment throught Credit card

data2=df.where(df['Payment']=='Credit card')
data2.dropna(inplace=True)
```

```python
amountByProductline=data2.groupby('Product line')[['Quantity', 'amount']].sum().reset_index()
amountByProductline
```

```python
c=np.arange(len(amountbyProductline['Product line']))
plt.xticks(c,amountbyProductline['Product line'],rotation=90)
plt.bar(c,amountbyProductline['Quantity'],width=0.5,label='Quantity',color='red')
plt.bar(c+0.3,amountbyProductline['amount'],width=0.5,label='Amount',color='black')
plt.xlabel('Product line',color='red')
plt.ylabel('Amount and Quantity',color='red')
plt.legend()
plt.show()
```

```python
data3=df.where(df['Payment']=='Ewallet')
data3.dropna(inplace=True)
```

```python
amountbyProductline=data3.groupby('Product line')[['Quantity', 'amount']].sum().reset_index()
amountbyProductline
```

```python
c=np.arange(len(amountbyProductline['Product line']))
plt.xticks(c,amountbyProductline['Product line'],rotation=90)
plt.bar(c,amountbyProductline['Quantity'],width=0.5,label='Quantity',color='red')
plt.bar(c+0.3,amountbyProductline['amount'],width=0.5,label='Amount',color='black')
plt.xlabel('Product line',color='red')
plt.ylabel('Amount and Quantity',color='red')
plt.legend()
plt.show()
```
